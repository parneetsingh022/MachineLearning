```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("data/home_data.csv")
TRAIN_SIZE = int(0.8*len(df))

train_df = df[0:TRAIN_SIZE]
test_df = df[TRAIN_SIZE:]
df.tail()
```

```python
comp = ['lat', 'long','bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'yr_built','condition', 'waterfront','view', 'grade','sqft_above', 'sqft_basement']

reg = linear_model.LinearRegression()

reg.fit(
    train_df[comp],
    train_df.price
)

reg.score(test_df[comp], test_df.price)
```
